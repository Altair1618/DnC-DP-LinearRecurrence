```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
A_DP = pd.read_csv('../data/A_DP.csv')
A_DNC = pd.read_csv('../data/A_DNC.csv')
```

```python
dataChart1 = pd.concat([A_DP.groupby('N')['Duration'].mean().reset_index(), A_DNC.groupby('N')['Duration'].mean().reset_index()], keys=['DP', 'DNC'])

line = sns.lineplot(x="N", y="Duration", marker="o", hue=dataChart1.index.get_level_values(0), data=dataChart1)

plt.title("Pengaruh N terhadap Durasi Rata-Rata")
plt.ylabel("Durasi Rata-Rata (ms)")
plt.show()
```

```python
dataChart2 = pd.concat([A_DP.groupby('N')['Peak Memory'].mean().reset_index(), A_DNC.groupby('N')['Peak Memory'].mean().reset_index()], keys=['DP', 'DNC'])
dataChart2['Peak Memory'] = dataChart2['Peak Memory'] / 1024

line = sns.lineplot(x="N", y="Peak Memory", marker="o", hue=dataChart2.index.get_level_values(0), data=dataChart2)

plt.title("Pengaruh N terhadap Pemakaian Memori")
plt.ylabel("Pemakaian Memori (KB)")
plt.show()
```
